# Chapter-007-009B-Keras-01

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import os
# fix random seed for reproducibility
np.random.seed(7)
```

```python
# load pima indians dataset
data_dir = os.path.abspath(os.path.join("..","..", "data"))
print(data_dir)
data_file = os.path.join(data_dir, "Sickness13.csv")
print(data_file)
```

```python
rawdata = np.loadtxt(data_file, delimiter=",")
print('Raw Data Set:', rawdata.shape)
dataset = rawdata[:]
print('Data Set:', dataset.shape)
```

```python
# split into input (X) and output (Y) variables
X = np.array(dataset[:,0:8], dtype=np.int32)
Y = np.array(dataset[:,8], dtype=np.int32)
print('X:', X.shape)
print('Y:', Y.shape)
```

```python
# create model
model = Sequential()

model.add(Dense(12, 
                input_dim=8, 
                activation='relu'
               )
         )

model.add(Dense(8, 
                activation='softmax'
               )
         )

model.add(Dense(1, 
                activation='sigmoid'
               )
         )
```

```python
# Compile model
model.compile(loss='binary_crossentropy', 
              optimizer='adam', 
              metrics=['accuracy']
             )
```

```python
# Fit the model
model.fit(X, 
          Y, 
          epochs=25, 
          batch_size=10,
          steps_per_epoch=250, 
          shuffle=True,
          workers=64,
          use_multiprocessing=True, 
          verbose=True
         )
```

```python
# evaluate the model
scores = model.evaluate(X, Y)

print("\n%s: %.5f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
# calculate predictions
predictions = model.predict(X)
#print(predictions.shape)
rounded = [round(x[0]) for x in predictions]

for i in range(predictions.shape[0]):
    p = rounded[i]
    r = Y[i]
    if p == r:
        ch='Yes'
    else:
        ch='No'
        
    print("%4.0f -> %2.4f => %2.0f => %2.0f # %s" % (i, predictions[i], p, r, ch))
```
