```python
import os
import numpy as np
import pandas as pd
import pandas_profiling as pdp
import webbrowser
```

```python
dirname = './990-DL/200-Structured-Zone/'

filenames=[
    '14-05-Assess-Fintech-01.csv',
    '14-05-Assess-Fintech-02.csv',
    '14-05-Assess-Fintech-03.csv',
    '14-05-Assess-Fintech-04.csv',
]
```

```python
%%javascript
IPython.OutputArea.auto_scroll_threshold = 9999;
```

```python
t=0
tmax=len(filenames)
for filename in filenames:
    t+=1
    print('#'*80)
    print('Profile Data Set: %d of %d' % (t, tmax))
    print('-'*60)
    print(filename)
    pathRealDataName=os.path.realpath(dirname)
    #print(pathRealDataName)
    fileFullDataName= os.path.join(pathRealDataName, filename)
    df= pd.read_csv(fileFullDataName, encoding='utf-8', compression=None, header=0)
    df.columns=['DateTime','EuroBuyAtUSD','EuroSellAtUSD','VolumeTrade']
    print('-'*60)
    print('Data Shape:', df.shape)
    print('-'*60)
    print('Describe Data')
    print('-'*60)
    print(df.describe())
    print('-'*60)
    print('Info Data')
    print('-'*60)
    print(df.info())
    print('-'*60)
    print('Profiling ....', filename)
    print('-'*60)
    profile = pdp.ProfileReport(df)
    rejected_variables = profile.get_rejected_variables(threshold=0.9)
    print('-'*60)
    print('Reject:', rejected_variables)
    print('-'*60)
    wwwfile='./990-DL/200-Structured-Zone/09-05-Utility-Data-Profiler-02-%02d.html' % (t)
    print('Creating profile:', wwwfile)
    print('-'*60)
    profile.to_file(outputfile=wwwfile)
    wwwRealfile=os.path.realpath(wwwfile)
    print('See:', wwwRealfile)
    webbrowser.open_new_tab(wwwRealfile)    
    print('#'*80,'\n','\n')
```
