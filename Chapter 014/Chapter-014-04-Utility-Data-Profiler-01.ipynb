```python
import os
import numpy as np
import pandas as pd
import pandas_profiling as pdp
```

```python
dirname = './990-DL/200-Structured-Zone/'
filename='14-01-Assess-Healthcare.csv'
fileZipname=filename + '.gz'
print(fileZipname)
```

```python
pathRealDataName=os.path.realpath(dirname)
print(pathRealDataName)
```

```python
fileFullDataName= os.path.join(pathRealDataName, fileZipname)
```

```python
df= pd.read_csv(fileFullDataName, encoding='utf-8', compression='gzip')
print(df.shape)
```

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
df.head()
```

```python
print(df.describe())
```

```python
print(df.info())
```

```python
profile = pdp.ProfileReport(df)
```

```python
%%javascript
IPython.OutputArea.auto_scroll_threshold = 9999;
```

```python
profile
```

```python
rejected_variables = profile.get_rejected_variables(threshold=0.9)
```

```python
print(rejected_variables)
```
