## Part B - Calculate Robot

```python
PP=True
```

```python
ValidGridList=[]
for tgridx in range(-10,100,1):
    for tgridy in range(-10,100,1):
        x=float(tgridx)
        y=float(tgridy)
        ValidGridList.append([x,y])
ValidGrid=np.array(ValidGridList)
print('Grid:',ValidGrid.shape)
```

```python
ValidRobotList=[]
ValidRobot=np.array(ValidRobotList)
InvalidRobotList=[]
t=0
e=0
s=0
for trobot in range(75,10,-1):
    s+=1       
    A=float(trobot)
    B=float(trobot)
    robot=True 
    for g in range(ValidGrid.shape[0]):
        r1=np.sqrt(np.power(x,2)+np.power(y,2))
        r2=A+B
        if r1 > r2:
            robot=False
        if robot==True:
            try:
                u=(np.power(x,2)+np.power(x,2)-np.power(A,2)-np.power(B,2))
                l=(2*A*B)
                P=np.arccos(u/l)
            except Exception: 
                robot=False
                pass
            if np.isnan(P):
               robot=False 

    if robot==True:            
        t+=1
        ValidRobotList.append([t, A, B])
        ValidRobot=np.array(ValidRobotList)
    else:
        e+=1
        InvalidRobotList.append([e, A, B])
        
ValidRobot=np.array(ValidRobotList)
print(ValidRobot.shape)
        
InvalidRobot=np.array(InvalidRobotList)
print(InvalidRobot.shape)

print('#' * 80)
print('Spawn %0d robots only %0d extinct with %0d was fit with survival rate of %0.3f persent' % (s, e, t, ((t/s)*100)))
print('#' * 80)
```

```python
for i in range(ValidRobot.shape[0]):
    rNr = ValidRobot[i][0]
    A=ValidRobot[i][1]
    B=ValidRobot[i][2]
    r=5
    basePrice=1000
    v=(np.pi*np.power(r,2)*A) + (np.pi*np.power(r,2)*B)
    armPrice=20*v
    TotalPrice=basePrice+armPrice
    
    print('%03d) Robot A=%03d B=%03d > Cost= £ %9.2f' % (rNr, A, B, TotalPrice))
```

```python
tmax=(ValidGrid.shape[0]-1)
tGrid=np.linspace(0,tmax,15)
print(len(tGrid), ValidGrid.shape[0])
```

```python
pcnt=0
for i in range(ValidRobot.shape[0]):
    for tp in tGrid:
        pcnt+=1
        t=int(tp)
        A=ValidRobot[i][1]
        B=ValidRobot[i][2]
        x=ValidGrid[t][0]
        y=ValidGrid[t][1]
        
        print('#' * 20)
        print('Robot A=%05d B=%05d for point x=%05d y=%05d' % (A, B, x, y))

        u=(np.power(x,2)+np.power(x,2)-np.power(A,2)-np.power(B,2))
        l=(2*A*B)
        P=np.arccos(u/l)
        if np.isnan(P):
            robot=False
        else:
            robot=True

        ukx=(-1*(B*np.sin(P)))
        uky=(A+(B*np.cos(P)))
        u=ukx*x + uky*y

        lkx=(A+(B*np.cos(P)))
        lky=(B*np.sin(P))
        l=lkx*x + lky*y

        Q=u/l

        x2=x
        y2=y

        Qd=np.degrees(Q)
        Pd=np.degrees(P)
        #print('Angles Q=%0.3f P=%0.3f' % (Qd,Pd))

        x0=0
        y0=0
        x1 = A * np.cos(Q)
        y1 = A * np.sin(Q)
        #print('x1=%0.3f y1=%0.3f x2=%0.3f y2=%0.3f x=%0.3f y=%0.3f' % (x1,y1,x2,y2,x,y))

        lmax = np.max([int(1.5*(A+B)),int(abs(x*2)),int(abs(y*2)),int(abs(x1*2)),int(abs(y1*2))])+100
        lcenX = int(4*(lmax/5))
        lcenY = int(lmax/2)

        py0=int(lcenX)
        px0=int(lcenY)

        py1=int(lcenX-x1)
        px1=int(lcenY+y1)

        py2=int(lcenX-x2)
        px2=int(lcenY+y2)

        py3=int(lcenX-x)
        px3=int(lcenY+y)

        lmaxY=int(lmax*1)
        lmaxX=int(lmax*1.5)

        #print(lmaxX ,lmaxY)
        #print(px0 ,py0, px1, py1)
        #print(px1 ,py1, px2, py2)
        #print(px3 ,py3)

        im = Image.new('RGBA', (lmaxX, lmaxY), (255, 255, 255, 255)) 
        draw = ImageDraw.Draw(im)
        if robot==True:
            tpstr='Valid Robot (%03d)' % (pcnt)
            draw.text((10, 10), tpstr, fill=(0,0,0,0))
            tpstr='(A=%0d B=%0d Q=%0.3f P=%0.3f)' % (A,B,Qd,Pd)
            draw.text((10, 20), tpstr, fill=(0,0,0,0))
        else:    
            tpstr='Invalid Robot (%03d)' % (pcnt)
            draw.text((10, 10), tpstr, fill=(0,0,0,0))  
            tpstr='(A=%0d B=%0d Q=%0.3f P=%0.3f)' % (A,B,Qd,Pd)
            draw.text((10, 20), tpstr, fill=(0,0,0,0))   
            tpstr='(Need Reach=%0.3f Only have=%0.3f)' % (r1,r2)
            draw.text((10, 30), tpstr, fill=(0,0,0,0))

        tpstr='X=%0d Y=%0d' % (x0,y0)
        draw.text((px0+15, py0), tpstr, fill=(0,0,0,0))

        tpstr='X=%0d Y=%0d' % (x1,y1)
        draw.text((px1-75, py1), tpstr, fill=(0,0,0,0))

        tpstr='X=%0d Y=%0d' % (x2,y2)
        draw.text((px2+15, py2), tpstr, fill=(0,0,0,0))

        tpstr='X=%0d Y=%0d' % (x,y)
        draw.text((px3+15, py3), tpstr, fill=(0,0,0,0))

        draw.line((px0 ,py0, px1, py1), fill=(255,0,0,0), width=3)
        draw.line((px1 ,py1, px2, py2), fill=(0,255,0,0), width=3)
        r=10
        draw.ellipse((px0-r, py0-r, px0+r, py0+r), fill=(0,0,0,0))
        r=2
        draw.ellipse((px1-r, py1-r, px1+r, py1+r), fill=(255,0,0,0))
        r=5
        draw.ellipse((px2-r, py2-r, px2+r, py2+r), fill=(0,255,0,0))
        r=2
        draw.ellipse((px3-r, py3-r, px3+r, py3+r), fill=(0,0,255,0))
        if PP==False:
            im.show()
        else:
            resultimage='../../Results/Chapter 12/Chapter-012-19-Evolutionary-01-%03d.bmp' % (pcnt)
            print(resultimage)
            im.save(resultimage)
```
