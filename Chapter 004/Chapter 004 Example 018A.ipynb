```python
%matplotlib inline
import os
import numpy as np
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt

# Create a random dataset
rng = np.random.RandomState(1)
X = np.sort(5 * rng.rand(80, 1), axis=0)
y = np.sin(X).ravel()
y[::5] += 3 * (0.5 - rng.rand(16))

# Fit regression model
regr_1 = DecisionTreeRegressor(max_depth=2)
regr_2 = DecisionTreeRegressor(max_depth=5)
regr_3 = DecisionTreeRegressor(max_depth=7)
regr_4 = DecisionTreeRegressor(max_depth=None)

regr_1.fit(X, y)
regr_2.fit(X, y)
regr_3.fit(X, y)
regr_4.fit(X, y)

# Predict
X_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]
y_1 = regr_1.predict(X_test)
y_2 = regr_2.predict(X_test)
y_3 = regr_3.predict(X_test)
y_4 = regr_4.predict(X_test)

# Plot the results
fig=plt.figure(figsize=(12, 6))
plt.scatter(X, y, s=20, edgecolor="black", c="darkorange", label="data", marker='D')
plt.plot(X_test, y_1, color="cornflowerblue", label="max_depth=2", linewidth=2)
plt.plot(X_test, y_2, color="yellowgreen", label="max_depth=5", linewidth=2)
plt.plot(X_test, y_3, color="aqua", label="max_depth=7", linewidth=2)
plt.plot(X_test, y_3, color="pink", label="max_depth=Max", linewidth=2)
plt.xlabel("data")
plt.ylabel("target")
plt.title("Decision Tree Regression")
plt.legend()
plt.show()
```

```python
print('Score max_depth=2  : %7.4f' % (regr_1.score(X, y)))
print('Score max_depth=5  : %7.4f' % (regr_2.score(X, y)))
print('Score max_depth=7  : %7.4f' % (regr_3.score(X, y)))
print('Score max_depth=Max: %7.4f' % (regr_4.score(X, y)))
```
