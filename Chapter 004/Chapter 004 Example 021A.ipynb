```python
from sklearn.neighbors.nearest_centroid import NearestCentroid
import numpy as np
import os
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X = np.array([[-2, -3], [-3, -2], [-2, -1], [-1, -1], [1, 1], [2, 1], [3, 2]])
y = np.array([0, 0, 0, 0, 1, 1, 1])
```

```python
clf1 = NearestCentroid(metric='euclidean', shrink_threshold=None)
clf1.fit(X, y)
```

```python
CX=clf1.centroids_
print(CX)
Cy=clf1.predict(CX)
print(Cy)
```

```python
t=0
for i in range(-35,40,2):
    for j in range(-35,40,2):
        xr=i/10
        yr=j/10
        pr=int(clf1.predict([[xr, yr]]))
        r=[xr,yr,pr]
        t+=1
        if t==1:
            Z=[r]
        else:
            Z=np.append(Z, [r], axis=0)
        #print('X=%5.2f Y=%5.2f p=%5.2f' % (xr,yr,pr))
```

```python
print(Z.shape)
```

```python
MLName='Nearest Centroid using Euclidean'
fig=plt.figure(figsize=(12, 12))
ax = fig.subplots(1, 1)
plt.title(MLName)

Z0, Z1, Z2 = Z[:, 0], Z[:, 1], Z[:, 2]
ax.scatter(Z0, Z1, c=Z2, cmap=plt.cm.coolwarm, s=250, edgecolors='k', marker='o', alpha=0.2)

X0, X1 = X[:, 0], X[:, 1]
ax.scatter(X0, X1, c=y, cmap=plt.cm.coolwarm, s=200, edgecolors='k', marker='P', alpha=1.0)

C0, C1 = CX[:, 0], CX[:, 1]
ax.scatter(C0, C1, c=Cy, cmap=plt.cm.coolwarm, s=800, edgecolors='k', marker='^', alpha=0.5)

plt.show()
```
