## Part A - Load libraries

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import f1_score
import numpy as np
import os
import pandas as pd
from random import random
import matplotlib.pyplot as plt
%matplotlib inline
```

## Part B - Load Data

```python
fileName = '../../Data/Cluster04.csv'
fileFullName = os.path.abspath(fileName)
print(fileFullName)
```

```python
clusterdf= pd.read_csv(fileFullName,header=0)
print(clusterdf.shape)
print(clusterdf.columns)
```

```python
X = np.array(clusterdf[['X', 'Y']], dtype='float64') 
y = np.array(clusterdf['T'], dtype='float64')
```

## Part C - Run Model

```python
clf = LinearDiscriminantAnalysis(solver='eigen', shrinkage='auto')
```

```python
MLName = 'Linear Discriminant Analysis'
```

```python
clf.fit(X, y)
```

```python
xdf=pd.DataFrame(X)
xdf.columns = (['X','Y'])

ydf=pd.DataFrame(y)
ydf.columns = (['T'])
Z1=pd.concat([xdf, ydf], axis=1)

transformer = RobustScaler(copy=True, with_centering=False, with_scaling=False, quantile_range=(0.1, 90.0)).fit(X)
Xp=transformer.transform(X)
xpdf=pd.DataFrame(X)
xpdf.columns = (['X','Y'])

yp=clf.predict(Xp)
ypdf=pd.DataFrame(yp)
ypdf.columns = (['T'])
Z2=pd.concat([xpdf, ypdf], axis=1)

Z=pd.concat([Z1, Z2], axis=0, sort=False)
Z.to_csv('../../Data/Cluster04.csv',index=False)

clusterdf= pd.read_csv(fileFullName,header=0)
print(clusterdf.shape)

X = np.array(clusterdf[['X', 'Y']], dtype='float64') 
y = np.array(clusterdf['T'], dtype='float64')
```

```python
def make_meshgrid(x, y, h=1):
    x_min, x_max = int(x.min()) - 2, int(x.max()) + 2
    y_min, y_max = int(y.min()) - 2, int(y.max()) + 2
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    return xx, yy
```

```python
def plot_contours(ax, clf, xx, yy, **params):
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    out = ax.contourf(xx, yy, Z, **params)
    return out
```

```python
fig = plt.figure(1, figsize=(12, 9))
ax = fig.subplots(1, 1)
plt.title(MLName)
X0, X1 = X[:, 0], X[:, 1]
xx, yy = make_meshgrid(X0, X1)
plot_contours(ax, clf, xx, yy, cmap=plt.cm.coolwarm, alpha=0.8)
ax.scatter(X0, X1, c=y, cmap=plt.cm.coolwarm, s=150, edgecolors='k', marker='*')
plt.show()
```

```python
p=[]
t=[]
for i in range(X.shape[0]):
    x=int(clf.predict(np.c_[X[i][0],X[i][1]]))
    p.append(x)
    t.append(y[i])
```

```python
f1score1=f1_score(t, p, average='micro')
f1score2=f1_score(t, p, average='macro')
```

```python
print('F1 Score (micro): %0.8f' % (f1score1))
print('F1 Score (macro): %0.8f' % (f1score2))
```
