## Part A - Load libraries

```python
from sklearn.metrics import f1_score
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

## Part B - Load Data

```python
fileName = '../../Data/Cluster02.csv'
fileFullName = os.path.abspath(fileName)
print(fileFullName)
```

```python
clusterdf= pd.read_csv(fileFullName,header=0)
print(clusterdf.shape)
print(clusterdf.columns)
```

```python
X = np.array(clusterdf[['X', 'Y']]) 
y = np.array(clusterdf['T'])
```

## Part C - Run Model

```python
clf = QuadraticDiscriminantAnalysis()
```

```python
MLName = 'Quadratic Discriminant Analysis'
```

```python
clf.fit(X, y)
```

```python
def make_meshgrid(x, y, h=1):
    x_min, x_max = int(x.min()) - 2, int(x.max()) + 2
    y_min, y_max = int(y.min()) - 2, int(y.max()) + 2
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    return xx, yy
```

```python
def plot_contours(ax, clf, xx, yy, **params):
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    out = ax.contourf(xx, yy, Z, **params)
    return out
```

```python
fig = plt.figure(1, figsize=(12, 9))
ax = fig.subplots(1, 1)
plt.title(MLName)
X0, X1 = X[:, 0], X[:, 1]
xx, yy = make_meshgrid(X0, X1)
plot_contours(ax, clf, xx, yy, cmap=plt.cm.coolwarm, alpha=0.8)
ax.scatter(X0, X1, c=y, cmap=plt.cm.coolwarm, s=100, edgecolors='k', marker='^')
plt.show()
```

```python
p=[]
t=[]
for i in range(X.shape[0]):
    x=int(clf.predict(np.c_[X[i][0],X[i][1]]))
    p.append(x)
    t.append(y[i])
```

```python
f1score=f1_score(t, p, average='macro')
```

```python
print('F1 Score: %0.4f' % (f1score))
```
