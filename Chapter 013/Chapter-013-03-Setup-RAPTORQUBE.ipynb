# Create Curated Zone - TPOLE Structure

![Data Lake Curated Zone](../../images/TPOLE.JPG)

```python
TPOLE=['Time','Person','Object','Location','Event']
```

## Hub

```python
dvpath='Hub'
for hubpath in TPOLE:
    dirname = os.path.join(pathDLname, dirpath, dvpath, hubpath)
    if not os.path.exists(dirname):
        os.makedirs(dirname)
        print('Make', dirname)
    else:
        print('Have:', dirname)    
```

## Satellite - Hub 

```python
dvpath='Satellite'
for hubSatellitepath in TPOLE:
    dirname = os.path.join(pathDLname, dirpath, dvpath, 'Hub', hubSatellitepath)
    if not os.path.exists(dirname):
        os.makedirs(dirname)
        print('Make', dirname)
    else:
        print('Have:', dirname)
```

## Links

#### 2 Links

```python
dvpaths=['Link','Satellite/Link']
for dvpath in dvpaths:
    for linkpath1 in TPOLE:
        for linkpath2 in TPOLE:
            linkpath = linkpath1 + '-' + linkpath2
            dirname = os.path.join(pathDLname, dirpath, dvpath, linkpath)
            if not os.path.exists(dirname):
                os.makedirs(dirname)
                print('Make:', dirname)
            else:
                print('Have:', dirname)
```

#### 3 Links

```python
dvpaths=['Link','Satellite/Link']
for dvpath in dvpaths:
    for linkpath1 in TPOLE:
        for linkpath2 in TPOLE:
            for linkpath3 in TPOLE:
                linkpath = linkpath1 + '-' + linkpath2  + '-' + linkpath3
                dirname = os.path.join(pathDLname, dirpath, dvpath, linkpath)
                if not os.path.exists(dirname):
                    os.makedirs(dirname)
                    print('Make:', dirname)
                else:
                    print('Have:', dirname)
```

#### 4 Links

```python
dvpaths=['Link','Satellite/Link']
for dvpath in dvpaths:
    for linkpath1 in TPOLE:
        for linkpath2 in TPOLE:
            for linkpath3 in TPOLE:
                for linkpath4 in TPOLE:
                    linkpath = linkpath1 + '-' + linkpath2  + '-' + linkpath3   + '-' + linkpath4
                    dirname = os.path.join(pathDLname, dirpath, dvpath, linkpath)
                    if not os.path.exists(dirname):
                        os.makedirs(dirname)
                        print('Make:', dirname)
                    else:
                        print('Have:', dirname)
```

#### 5 Links

```python
dvpaths=['Link','Satellite/Link']
for dvpath in dvpaths:
    for linkpath1 in TPOLE:
        for linkpath2 in TPOLE:
            for linkpath3 in TPOLE:
                for linkpath4 in TPOLE:
                    for linkpath5 in TPOLE:
                        linkpath = linkpath1 + '-' + linkpath2  + '-' + linkpath3 + '-' + linkpath4  + '-' + linkpath5
                        dirname = os.path.join(pathDLname, dirpath, dvpath, linkpath)
                        if not os.path.exists(dirname):
                            os.makedirs(dirname)
                            print('Make:', dirname)
                        else:
                            print('Have:', dirname)
```
