```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
print(cv.__version__)
```

```python
imgfile = '../../data/dog1.png'
img = cv.imread(imgfile, cv.IMREAD_UNCHANGED)
```

```python
imgdim = img.shape
imgheight = img.shape[0] 
imgwidth = img.shape[1] 
imgchannels = img.shape[2] 
 
print('Image Dimension    : ',imgdim) 
print('Image Height       : ',imgheight) 
print('Image Width        : ',imgwidth) 
print('Number of Channels : ',imgchannels) 
```

```python
B,G,R,A = cv.split(img)
```

```python
print('B:', B.shape)
print('G:', G.shape)
print('R:', R.shape)
print('A:', A.shape)
```

```python
if len(img.shape) > 2 and img.shape[2] == 4:
    #convert the image from RGBA2RGB
    img2 = cv.cvtColor(img, cv.COLOR_BGRA2BGR)
    img3 = cv.cvtColor(img2, cv.COLOR_BGR2HSV)
    img4 = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
imgdim = img2.shape
imgheight = img2.shape[0] 
imgwidth = img2.shape[1] 
imgchannels = img2.shape[2] 
 
print('Image Dimension    : ',imgdim) 
print('Image Height       : ',imgheight) 
print('Image Width        : ',imgwidth) 
print('Number of Channels : ',imgchannels) 
```

```python
imgdim = img3.shape
imgheight = img3.shape[0] 
imgwidth = img3.shape[1] 
imgchannels = img3.shape[2] 
 
print('Image Dimension    : ',imgdim) 
print('Image Height       : ',imgheight) 
print('Image Width        : ',imgwidth) 
print('Number of Channels : ',imgchannels) 
```

```python
imgdim = img4.shape
imgheight = img4.shape[0] 
imgwidth = img4.shape[1]
if len(img4.shape) > 2:
    imgchannels = img4.shape[2]
else:
    imgchannels=0
 
print('Image Dimension    : ',imgdim) 
print('Image Height       : ',imgheight) 
print('Image Width        : ',imgwidth) 
print('Number of Channels : ',imgchannels) 
```

```python
imgd=img4
h=imgd.shape[0]
w=imgd.shape[1]
plt.figure(figsize=(h/100,w/100))
titlestr='%d %% Dog - (%d x %d)' % (100,h,w)
plt.title(titlestr)
plt.imshow(imgd,aspect='auto', cmap = 'gray')
plt.axis('off')
plt.show()
```

```python
scale_percent = 30 # percent of original size
width = int(img.shape[1] * scale_percent / 100) 
height = int(img.shape[0] * scale_percent / 100) 
dim = (width, height) 

# resize image
resized30 = cv.resize(img, dim, interpolation = cv.INTER_AREA) 
 
print('%d%% Resized Dimensions : (%d, %d)' % (scale_percent,resized30.shape[0],resized30.shape[1]))
```

```python
imgd=resized30
h=imgd.shape[0]
w=imgd.shape[1]
plt.figure(figsize=(h/100,w/100))
titlestr='%d%% Dog - (%d x %d)' % (scale_percent,h,w)
plt.title(titlestr)
plt.imshow(imgd,aspect='auto',cmap='autumn')
plt.axis('off')
plt.show()
```

```python
scale_percent = 10 # percent of original size
width = int(img.shape[1] * scale_percent / 100) 
height = int(img.shape[0] * scale_percent / 100) 
dim = (width, height) 

# resize image
resized10 = cv.resize(img, dim, interpolation = cv.INTER_AREA) 
 
print('%d%% Resized Dimensions : (%d, %d)' % (scale_percent,resized10.shape[0],resized10.shape[1]))
```

```python
imgd=resized10
h=imgd.shape[0]
w=imgd.shape[1]
plt.figure(figsize=(h/100,w/100))
titlestr='%d%% Dog - (%d x %d)' % (scale_percent,h,w)
plt.title(titlestr)
plt.imshow(imgd,aspect='auto',cmap='autumn')
plt.axis('off')
plt.show()
```

```python
scale_percent = 5 # percent of original size
width = int(img.shape[1] * scale_percent / 100) 
height = int(img.shape[0] * scale_percent / 100) 
dim = (width, height) 

# resize image
resized = cv.resize(img, dim, interpolation = cv.INTER_AREA) 
 
print('%d%% Resized Dimensions : (%d, %d, %d)' % (scale_percent,resized.shape[0],resized.shape[1],resized.shape[2]))
```

```python
imgd=resized
h=imgd.shape[0]
w=imgd.shape[1]
plt.figure(figsize=(h/100,w/100))
titlestr='%d%% Dog - (%d x %d)' % (scale_percent,h,w)
plt.title(titlestr)
plt.imshow(imgd,aspect='auto',cmap='autumn')
plt.axis('off')
plt.show()
```
