## Chapter-011-19-Electronics-01

```python
for a in range (2):
    for b in range (2):
        x= a and b
        print(' %0d and %0d => %0d' % (a, b, x))
```

```python
for a in range (2):
    for b in range (2):
        x= a or b
        print(' %0d and %0d => %0d' % (a, b, x))
```

```python
for a in range (2):
        x= not a
        print(' %0d => %0d' % (a, x))
```

```python
t=0
for p1 in range (2):
    for p2 in range (2):
        for p3 in range (2):
            for p4 in range (2):
                a1=p1
                b1=p2
                a2=p3
                b2=p4
                t+=1
                print('====================================')
                print('c= %0d:p1=%0d p2=%0d p3=%0d p4=%0d' % (t, p1, p2, p3, p4))
                print('====================================')
                x1=a1 and b1
                print('%0d and %0d => %0d' % (a1,b1,x1))
                                                                                            
                x2=a2 and b2
                print('%0d and %0d => %0d' % (a2,b2,x2))

                a3 = x1
                b3 = x2

                x3=a3 or b3
                print('%0d or %0d => %0d' % (a3,b3,x3))

                a4 = x3

                x4= not a4
                print('not %0d => %0d' % (a4, x4))

                p5=a4
                print('====================================')
                print('p5=%0d' % (p5))
                print('====================================')
```
