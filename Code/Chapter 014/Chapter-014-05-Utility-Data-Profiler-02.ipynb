```python
import os
import numpy as np
import pandas as pd
import pandas_profiling as pdp
import webbrowser
```

```python
dirname = os.path.abspath('../../Results/Chapter 14/990-DL/200-Structured-Zone/')
print(dirname)
```

```python
%%javascript
IPython.OutputArea.auto_scroll_threshold = 9999;
```

```python
wwwDir=os.path.abspath('../../Results/Chapter 14/990-DL/200-Structured-Zone/')
```

```python
tMax=10
for t in range(1,tMax,1):
    fileOutname='14-05-Access-Fintech-%02d.csv' % t
    fullRealName = os.path.abspath(os.path.join(dirname, fileOutname)) 
    print('#'*80)
    print('Profile Data Set: %d of %d' % (t, (tMax-1)))
    print('-'*60)
    print(fileOutname)
    df=pd.read_csv(fullRealName)
    df.columns=['DateTime','EuroBuyAtUSD','EuroSellAtUSD','VolumeTrade']
    print('-'*60)
    print('Data Shape:', df.shape)
    print('-'*60)
    print('Describe Data')
    print('-'*60)
    print(df.describe())
    print('-'*60)
    print('Info Data')
    print('-'*60)
    print(df.info())
    print('-'*60)
    print('Profiling ....', fileOutname)
    print('-'*60)
    profile = pdp.ProfileReport(df)
    rejected_variables = profile.get_rejected_variables(threshold=0.9)
    print('-'*60)
    print('Reject:', rejected_variables)
    print('-'*60)
    wwwfileName='14-05-Utility-Data-Profiler-02-%02d.html' % (t)
    wwwfile=os.path.abspath(os.path.join(wwwDir, wwwfileName))
    print('Creating profile:', wwwfileName)
    print('-'*60)
    profile.to_file(outputfile=wwwfile)
    wwwRealfile=os.path.abspath(wwwfile)
    print('See:', wwwRealfile)
    webbrowser.open_new_tab(wwwRealfile)    
    print('#'*80,'\n','\n')
```
