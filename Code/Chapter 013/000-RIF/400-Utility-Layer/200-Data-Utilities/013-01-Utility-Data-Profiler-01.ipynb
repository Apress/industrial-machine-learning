```python
import numpy as np
import pandas as pd
import pandas_profiling as pdp
import string

rowcnt=10
columncnt=4
columnset=list(string.ascii_uppercase)
columnheader=columnset[0:columncnt]
print(columnheader)
```

```python
df=pd.DataFrame(
    np.random.rand(rowcnt, columncnt),
    columns=columnheader
)
```

```python
print(df.shape)
```

```python
print(df.describe())
```

```python
print(df.info())
```

```python
profile = pdp.ProfileReport(df)
```

```python
%%javascript
IPython.OutputArea.auto_scroll_threshold = 9999;
```

```python
profile
```

```python
rejected_variables = profile.get_rejected_variables(threshold=0.9)
```

```python
print(rejected_variables)
```

```python
wwwfile='./100-Data-Profiles/08-Utility-Data-Profiler-01.html'
profile.to_file(outputfile=wwwfile)
```
