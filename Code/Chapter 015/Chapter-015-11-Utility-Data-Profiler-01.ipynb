```python
import os
import numpy as np
import pandas as pd
import pandas_profiling as pdp
```

```python
dirname = './999-DL/100-Raw-Zone/'
filebase='15-04-Retrieve-Mars-Hopper'
filename=filebase + '.csv'
fileZipname=filename + '.gz'
print(fileZipname)
```

```python
pathRealDataName=os.path.realpath(dirname)
print(pathRealDataName)
```

```python
fileFullDataName= os.path.join(pathRealDataName, fileZipname)
```

```python
print(fileFullDataName)
```

```python
df= pd.read_csv(fileFullDataName, encoding='utf-8', compression='gzip')
print(df.shape)
```

```python
print(df.describe())
```

```python
print(df.info())
```

```python
df2=df.sample(frac=0.01, replace=False)
```

```python
print(df2.info())
```

```python
profile = pdp.ProfileReport(df2)
```
