k-means clustering is a method of vector quantization, originally from signal processing, that is popular for cluster analysis in data mining. 

k-means clustering aims to partition n observations into k clusters in which each observation belongs to the cluster with the nearest mean, serving as a prototype of the cluster. 

This results in a partitioning of the data space into Voronoi cells.

```python
N=int(9)
M=int(25)
```

```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
seednum=0
np.random.seed(seednum)
```

```python
from sklearn.cluster import KMeans
```

```python
K1=[]
for i in range(-M, M):
    K1 = K1+[[i,i]]
    K1 = K1+[[0,i]]
    K1 = K1+[[i,0]]
    K1 = K1+[[-i,i]]
#print(K1)
```

```python
X = np.array(K1)
```

```python
K2=[]
for i in range(-M, M):
    for j in range(-M, M):
        K2 = K2+[[i,j]]
#print(K2)
```

```python
T = np.array(K2)
```

## KMeans One

```python
km1 = KMeans(n_clusters=N, 
                random_state=seednum, 
                init='k-means++',
                algorithm='auto',
                tol=1e-06,
                max_iter=1000,
                n_jobs=-1,
                n_init=100,
                verbose=True
               )
```

```python
km1.fit(X)
```

```python
print(km1.get_params(deep=True))
```

```python
print(km1.n_iter_)
```

```python
print(km1.labels_)
```

```python
z1 = km1.predict(X)
```

```python
p1 = km1.predict(T)
```

## KMeans Two

```python
km2 = KMeans(n_clusters=N, 
                random_state=seednum, 
                init='random',
                algorithm='auto',
                tol=1e-06,
                max_iter=1000,
                n_jobs=-1,
                n_init=100,
                verbose=True
               )
```

```python
km2.fit(X)
```

```python
print(km2.get_params(deep=True))
```

```python
print(km2.n_iter_)
```

```python
print(km2.labels_)
```

```python
z2 = km2.predict(X)
```

```python
p2 = km2.predict(T)
```

```python
g1=pd.DataFrame(c2)
g1.columns=['Cluster']
gc1=g1['Cluster'].value_counts()
for i in range(gc1.shape[0]):
    print('Cluster %d : %d' % (i+1,gc1[i]))
print('----------------------')
print('Cluster   : %d' % (gc1.sum()))
```
