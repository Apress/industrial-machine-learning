## Chapter-009-008-Travelling-Salesman-01

```python
import random
import numpy
import math
import copy
import os

import matplotlib
matplotlib.use('TkAgg')
%matplotlib inline

import matplotlib.pyplot as plt
```

```python
citycnt = 20
cities = [random.sample(range(100), 2) for x in range(citycnt)];
tour = random.sample(range(citycnt),citycnt);
for temperature in numpy.logspace(0,5,num=100000)[::-1]:
    [i,j] = sorted(random.sample(range(citycnt),2));
    newTour =  tour[:i] + tour[j:j+1] +  tour[i+1:j] + tour[i:i+1] + tour[j+1:];
    if math.exp( ( sum([ math.sqrt(sum([(cities[tour[(k+1) % citycnt]][d] - cities[tour[k % citycnt]][d])**2 for d in [0,1] ])) for k in [j,j-1,i,i-1]]) - sum([ math.sqrt(sum([(cities[newTour[(k+1) % citycnt]][d] - cities[newTour[k % citycnt]][d])**2 for d in [0,1] ])) for k in [j,j-1,i,i-1]])) / temperature) > random.random():
        tour = copy.copy(newTour);
        
fig1 = plt.figure(1, figsize=(12, 9))
plt.plot([cities[tour[i % citycnt]][0] for i in range(citycnt+1)], [cities[tour[i % citycnt]][1] for i in range(citycnt+1)], 'xb-');
plt.show()
```
