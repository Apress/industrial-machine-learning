```python
import networkx as nx
import matplotlib.pyplot as plt
import os
```

```python
imagepath = os.path.join(*[os.path.dirname(os.path.dirname(os.getcwd())),'Results','Chapter 09'])
print(imagepath)

if not os.path.exists(imagepath):
    os.makedirs(imagepath)
```

```python
G1=nx.Graph()
edgelist = [
    ("a", "b", 10),
    ("a", "c", 11),
    ("a", "f", 20),
    ("b", "c", 14),
    ("b", "d", 17),
    ("c", "d", 19),
    ("c", "f", 11),
    ("d", "e", 10),
    ("e", "f", 30)
    ]
G1.add_weighted_edges_from(edgelist)
```

```python
fig1=plt.figure(figsize=(10, 10))

nx.draw(G1,pos=nx.spectral_layout(G1), nodecolor='green', edge_color='blue', node_size=850, node_shape='*', aplha=0.8)

nx.draw_networkx_labels(G1,pos=nx.spectral_layout(G1))

nx.draw_networkx_edge_labels(G1,pos=nx.spectral_layout(G1))

plt.show()
```

```python
print(nx.info(G1))
```

```python
print('Dijkstra path:')
print(nx.dijkstra_path(G1,'a','e'))
```

```python
imagename = os.path.join(*[os.path.dirname(os.path.dirname(os.getcwd())),'Results','Chapter 09','Chapter-009-004-Dijkstra-algorithm-02-01.jpg'])
fig1.savefig(imagename, bbox_inches='tight') 
```

```python
G2 = nx.cycle_graph(24)
pos = nx.spring_layout(G2, iterations=300)
fig2=plt.figure(figsize=(10, 10))
nx.draw(G2, pos, node_color=range(24), node_size=800, cmap=plt.cm.Greens)
plt.show()
```

```python
imagename = os.path.join(*[os.path.dirname(os.path.dirname(os.getcwd())),'Results','Chapter 09','Chapter-009-004-Dijkstra-algorithm-02-02.jpg'])
fig2.savefig(imagename, bbox_inches='tight') 
```

```python
t=0
for m in range(2,9):
    for n in range(2,9):
        print('-'*80)
        print(m, n)
        print('-'*80)
        G3 = nx.grid_2d_graph(m, n) 
        for line in nx.generate_adjlist(G3):
            print(line)

        graphlist='../../Results/Chapter 09/grid-%03d-%03d.edgelist' % (m, n)

        nx.write_edgelist(G3, path=graphlist, delimiter=":")

        H = nx.read_edgelist(path=graphlist, delimiter=":")

        fig3=plt.figure(figsize=(10, 10))

        nx.draw(H, node_size=80, node_color='green', alpha=0.8, cmap=plt.cm.Greens)

        plt.show()
        
        t=+1

        figname='Chapter-009-004-Dijkstra-algorithm-02-03-%02d-%02d-%02d.jpg' % (t,m, n)
        imagename = os.path.join(*[os.path.dirname(os.path.dirname(os.getcwd())),'Results','Chapter 09',figname])
        fig3.savefig(imagename, bbox_inches='tight') 
```
