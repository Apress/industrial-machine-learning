```python
import os
import numpy as np
import pandas as pd
import pandas_profiling as pdp
```

```python
dirDataName = '../../../../Chapter 08/990-DL/100-Raw-Zone/'
pathRealDataName=os.path.realpath(dirDataName)
print(pathRealDataName)
```

```python
fileFullDataName= os.path.join(pathRealDataName, fileDataName)
print(fileFullDataName)
```

```python
personFullDF= pd.read_csv(fileFullDataName, encoding='utf-8', compression='gzip')
print(personFullDF.shape)
```

```python
print(personFullDF.columns)
```

```python
print(personFullDF.describe())
```

```python
personFullDF.info()
```

```python
profile = pdp.ProfileReport(personFullDF)
```

```python
%%javascript
IPython.OutputArea.auto_scroll_threshold = 9999;
```

```python
profile
```

```python
rejected_variables = profile.get_rejected_variables(threshold=0.9)
```

```python
print(rejected_variables)
```

```python
profile.to_file(outputfile=wwwfile)
```
