## Chapter-010-17-Konigsberg-01

```python
import matplotlib
import numpy as np
matplotlib.use('TkAgg')
%matplotlib inline

import networkx as nx
import matplotlib.pyplot as plt
```

```python
nodes = [0, 3]
edges = [
    (0, 1, {'key': 'Grüne Brücke', 'color': 'blue', 'width': 5}),
    (0, 1, {'key': 'Köttelbrücke', 'color': 'red', 'width': 2}),
    (1, 2, {'key': 'Krämerbrücke', 'color': 'blue', 'width': 5}),
    (1, 2, {'key': 'Schmiedebrücke', 'color': 'red', 'width': 2}),
    (2, 3, {'key': 'Holzbrücke', 'color': 'blue', 'width': 5}),
    (3, 1, {'key': 'Dombrücke', 'color': 'blue', 'width': 5}),
    (3, 0, {'key': 'Hohe Brücke', 'color': 'blue', 'width': 5})
]
```

```python
# Create the graph
graph = nx.MultiGraph()
graph.add_nodes_from(nodes)
r=graph.add_edges_from(edges)
```

```python
for i, j, label in graph.edges(data='key'):
    print(i, j, label)
```

```python
labels = {}
for i, j, label in graph.edges(data='key'):
    if (i, j) not in labels:
        labels[(i, j)] = label
    else:
        labels[(i, j)] += '\n' + label
```

```python
colors = [x for _, _, x in graph.edges(data='color')]
```

```python
widths = [x for _, _, x in graph.edges(data='width')]
```

```python
pos = nx.spring_layout(graph,scale=2.5,iterations=250)
```

```python
# Plot the graph
%matplotlib inline
fig=plt.figure(figsize=(10,10))
nx.draw_networkx(graph, pos, edge_color=colors, with_labels=True, width=widths, alpha=1.0)
nx.draw_networkx_edge_labels(graph, pos, edge_labels=labels)
plt.axis('off')
plt.show()
```
