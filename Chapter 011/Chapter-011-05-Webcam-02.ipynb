# Chapter-011-05-Webcam-02

```python
import matplotlib
matplotlib.use('TkAgg')

import os
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
imgfile = '../../data/dog.png'
img = cv.imread(imgfile,0)
```

```python
imagepath = os.path.join(*[os.path.dirname(os.path.dirname(os.getcwd())),'Results','Chapter 11'])
print(imagepath)
if not os.path.exists(imagepath):
    os.makedirs(imagepath)
imgoutfile0 = os.path.join(*[os.path.dirname(os.path.dirname(os.getcwd())),'Results','Chapter 11','Chapter-11-05-Webcam-02-00.png'])
print(imgoutfile0)
imgoutfile1 = os.path.join(*[os.path.dirname(os.path.dirname(os.getcwd())),'Results','Chapter 11','Chapter-11-05-Webcam-02-01.png'])
print(imgoutfile1)
```

```python
cv.imwrite(imgoutfile0, img)
edges = cv.Canny(img,5,10)
imginv = cv.bitwise_not(edges)
cv.imwrite(imgoutfile1, imginv)
```

```python
imgd=img
h=imgd.shape[0]
w=imgd.shape[1]
plt.figure(figsize=(h/100,w/100))

plt.subplot(121)
plt.imshow(img, cmap = 'gray')
plt.title('Original Dog Image')
plt.xticks([]), plt.yticks([])
plt.subplot(122)
plt.imshow(imginv,cmap = 'gray')
plt.title('Edge Robot Image'), plt.xticks([]), plt.yticks([])
plt.show()
```
